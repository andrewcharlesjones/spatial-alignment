```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.io import mmread
import pandas as pd
from os.path import join as pjoin
```

```python
DATA_DIR = "./data/mouse_brain_slideseq/12_allMTXs_CCF"
metadata = pd.read_table(pjoin(DATA_DIR, "MBASS_d123_metadata.tsv"))
data_sparse = mmread(pjoin(DATA_DIR, "MBASS_d123_matrix.mtx"))
```

```python
data = data_sparse.toarray()
```
