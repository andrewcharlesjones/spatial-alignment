```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from os.path import join as pjoin
from scipy.io import mmread
import scanpy as sc
import anndata
import squidpy as sq
from gpsa import VariationalGPSA, rbf_kernel
import torch
from IPython.display import display, clear_output
```

```python
DATA_DIR = "../../../data/ben"
```

## Load data

```python
data1 = pd.read_csv(pjoin(DATA_DIR, "cell_table_immune.csv"))
data2 = pd.read_csv(pjoin(DATA_DIR, "cell_table_tumor.csv"))

# Spatial coordinates for immune cells
X_slice1 = data1[["centroid-0", "centroid-1"]].values

# Spatial coordinates for tumor cells
X_slice2 = data2[["centroid-0", "centroid-1"]].values

# Get overlapping marker names
marker_col_idx = np.arange(1, 47)
overlapping_markers = np.intersect1d(data1.columns.values[marker_col_idx], data2.columns.values[marker_col_idx])

# Marker data for immune cells
Y_slice1 = data1[overlapping_markers].values
Y_slice1 = np.log(Y_slice1 + 1)
Y_slice1 = (Y_slice1 - Y_slice1.mean(0)) / Y_slice1.std(0)

# Marker data for tumor cells
Y_slice2 = data2[overlapping_markers].values
Y_slice2 = np.log(Y_slice2 + 1)
Y_slice2 = (Y_slice2 - Y_slice2.mean(0)) / Y_slice2.std(0)
```

## Manually rotate one slice so they have same orientation

```python
plt.figure(figsize=(10, 10))
plt.subplot(221)
plt.scatter(X_slice1[:, 0], X_slice1[:, 1], s=10, c="black")
plt.title("Immune")
# plt.axis("off")

plt.subplot(222)
plt.scatter(X_slice2[:, 0], X_slice2[:, 1], s=10, c="black")
plt.title("Tumor")
plt.show()
```

## Filter for spots inside the tissue and create AnnData objects

```python
anndata_slice1 = anndata.AnnData(Y_slice1)
anndata_slice1.obsm['spatial'] = X_slice1

anndata_slice2 = anndata.AnnData(Y_slice2)
anndata_slice2.obsm['spatial'] = X_slice2
```

## Compute spatial autocorrelation (Moran's I) for each marker

```python
sq.gr.spatial_neighbors(anndata_slice1)
sq.gr.spatial_autocorr(
    anndata_slice1,
    mode="moran",
)

sq.gr.spatial_neighbors(anndata_slice2)
sq.gr.spatial_autocorr(
    anndata_slice2,
    mode="moran",
)
```
